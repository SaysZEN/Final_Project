```python
pop.dropna(inplace=True)
new_column_names = {'Indicator Name': 'year'}
pop.rename(columns=new_column_names, inplace=True)
pop_age_dis = pop[["year","Population ages 15-64 (% of total population)",
                                         "Population ages 0-14 (% of total population)",
                                         "Population ages 65 and above (% of total population)",
                                         "Population ages 0-14, total",
                                         "Population ages 15-64, total",
                                         "Population ages 65 and above, total",
                                         "Population growth (annual %)",
                                         "Population, total"]]
names = {"Population ages 15-64 (% of total population)": "Age_15-64_per",
         "Population ages 0-14 (% of total population)": "Age_0-14_per",
         "Population ages 65 and above (% of total population)": "Age_65+_per",
         "Population ages 0-14, total": "Age_0-14",
         "Population ages 15-64, total": "Age_15-64",
         "Population ages 65 and above, total": "Age_65+",
         "Population growth (annual %)": "Pop_grow",
         "Population, total": "Pop_tot"}
pop_age_dis = pop_age_dis.rename(columns = names)
pop_age_dis.set_index('year', inplace=True)
pop_age_dis_61_5 = pop_age_dis.loc[range(1961, 2023,5)]
pop_age_dis_61_5
```

```python
from pyecharts.charts import Pie
from pyecharts.charts import Bar
from pyecharts import options as opts
from IPython.display import IFrame
from pyecharts.charts import Line
bar = (
    Bar()
    .add_xaxis(["2005", "2010", "2015", "2020"])
    .add_yaxis("Age_0-14", [17.4, 16.9, 16.1, 15.0])
    .add_yaxis("Age_14-65", [84.0, 80.9, 76.3, 73.9])
    .add_yaxis("Age_65+", [26.3, 30.2, 34.7, 37.4])
     .set_global_opts(
        yaxis_opts=opts.AxisOpts(name="Population (M)", name_location="middle", name_gap=30),
        title_opts=opts.TitleOpts(title="Age_Structure")
))
bar.render("Age_str.html")
IFrame(src = 'Age_str.html', width=1000, height=600)
```

```python
#1961-2022
grow = pop_age_dis_61_5["Pop_grow"].values
grow = [f"{i:.3f}" for i in grow]
years = list(range(1961, 2023,5))  # 创建包含 1961 年到 2022 年的年份列表
years_str = str(years)
c = (
    Line()
    .add_xaxis(["1961","1966","1971","1976","1981","1986","1991","2001","2006","2011","2016","2021"])  # 使用年份列表作为 x 轴数据
    .add_yaxis(
        "Population_growth",
        grow,
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_="min"),
                opts.MarkPointItem(type_="max")
            ]
        ),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Population_growth"),
        xaxis_opts=opts.AxisOpts(interval=5)# 设置坐标轴间隔，每隔 5 年显示一个年份
    )
    .render("line_markpoint.html")
)
IFrame(src = 'line_markpoint.html', width=1000, height=520)
```

```python
x_data = ["1961","1966","1971","1976","1981","1986","1991","2001","2006","2011","2016","2021"]
Work_Pop = pop_age_dis_61_5["Age_15-64"].values

(
    Line()
    .add_xaxis(x_data)
    .add_yaxis(
        series_name="Working_Population",
        stack="总量",
        y_axis= [61.1,68.1,73.1,76.5,79.5,83.3,86.5,87.0,85.7,83.4,80.4,75.6,73.4],
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=True, position="top"),
    )
    .add_yaxis(
        series_name="Total_Population",
        stack="总量",
        y_axis=[94.0,98.9,105.7,112.8,117.7,121.5,124.0,125.8,127.1,127.9,127.8,127.0,125.7],
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=True, position="top"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Working population structure"),
        tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
    )
    .render("stacked_area_chart.html")
)
IFrame(src = 'stacked_area_chart.html', width=1000, height=520)
```
